```python
import pandas as pd
```

```python
data=pd.read_csv('one-hot.csv')
```

```python
data
```

```python
pd.get_dummies(data['Gender'])
```

```python
pd.get_dummies(data['Grade'])
```

### Lable encoding

```python
dict1={'female':0,'male':1}
data['label_gender']=data['Gender'].map(dict1)
```

```python
data[['Gender','label_gender']]
```

```python
dict2={'A':0,'B':1,'C':2}
data['label_grade']=data['Grade'].map(dict2)
```

```python
data[['Grade','label_grade']]
```

### frequency encoding

```python
dict3=data['Gender'].value_counts().to_dict()
```

```python
dict4=data['Grade'].value_counts().to_dict()
```

```python
data['fre_grade']=data['Grade'].map(dict4)
data['fre_gender']=data['Gender'].map(dict3)
```

```python
data[['Grade','fre_grade']]
```

```python
data[['Gender','fre_gender']]
```
