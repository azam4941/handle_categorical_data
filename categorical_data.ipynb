```python
import pandas as pd
```

```python
data=pd.read_csv('one-hot.csv')
```

```python
data
```

```python
pd.get_dummies(data['Gender'])
```

```python
pd.get_dummies(data['Grade'])
```

```python
dic1={'female':0,'male':1}
data['new_gender']=data['Gender'].map(dic1)
data[['Gender','new_gender']]
```

```python
dic2={'A':0,'B':1,'C':2}
data['new_grade']=data['Grade'].map(dic2)
data[['Grade','new_grade']]
```

```python
dict1=data['Gender'].value_counts().to_dict()
data['fre_gender']=data['Gender'].map(dict1)
data[['Gender','fre_gender']]
```

```python
dict2=data['Grade'].value_counts().to_dict()
data['fre_grade']=data['Grade'].map(dict2)
data[['Grade','fre_grade']]
```
